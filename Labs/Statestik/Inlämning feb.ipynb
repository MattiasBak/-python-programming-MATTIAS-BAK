Detta är ett väldigt rudimentärt analytiskt program som arbetat med statestik och linjär regrition.
Det som är viktigt är inte vad som syns här utan det som är under huven. Men det vi kan se hör är att den inte bara läser in den tilldelade data.

```python
from Labb_statestik import LinearRegression
import numpy as np

data = np.genfromtxt ("C:\\Users\\matti\\Downloads\\housing.csv", delimiter=',',skip_header=1, dtype=str)

X_raw = data [:, [2,3,4]]
y_raw = data[:,8]

```

```python
mask = ((X_raw !='').all(axis=1) & (y_raw !=''))
X = X_raw[mask].astype(float)
y = y_raw[mask].astype(float)
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
model.summary()
```

Utan den kan även tränas med data. Här kan vi se resultatet av att vi använder tre olika värden och gav den R^2 ≈ 0.089. vilket då innebär att den förklarar ca 8.9% av variationen i responsvariablen.
När vi kollar F- testet ser vi att p-värdenra är toklåga och kan därgenom uppskatta att alla bär en hög statistisk signifikans. Något som kan tala emot våran upptäckt är att förklarings graden är låg. Detta kan bero på ett antal saker, så som bristande data eg. variabler som saknas eller att det inte är statistiskt linjärt.  Modellen kan alltså inte anses vara bra för prediktion men visar ändå att vissa variabler påverkar utfallet.
